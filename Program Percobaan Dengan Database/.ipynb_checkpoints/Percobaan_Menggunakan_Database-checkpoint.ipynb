```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import train_test_split

active_gigs = data_gigs[data_gigs['is_active']==1]
#active_gigs
def content_based_recomendation(user_id,num_recomendation):
    active_gigs = data_gigs[data_gigs['is_active']==1]
    combined_data = pd.concat([data_view_gigs_history, data_wishlist, data_rating])
    reader = Reader(rating_scale=(1, 5))
    data = Dataset.load_from_df(combined_data[['user_id', 'gig_id', 'rating']], reader)
    trainset, testset = train_test_split(data, test_size=0.2)
    model = KNNBasic()
    model.fit(trainset)
    user_interacted_gigs = combined_data[combined_data['user_id'] == user_id]['gig_id']
    gigs_to_predict = data_gigs_images[~data_gigs_images['gig_id'].isin(user_interacted_gigs)]
    predicted_ratings = []
    for gig_id in gigs_to_predict['gig_id']:
        predicted_rating = model.predict(user_id, gig_id).est
        predicted_ratings.append((gig_id, predicted_rating))
    predicted_ratings.sort(key=lambda x: x[1], reverse=True)
    recommended_gigs = data_gigs_images.loc[data_gigs_images['gig_id'].isin([gig_id for gig_id, _ in predicted_ratings[:num_recommendations]])]
    recommended_gigs = recommended_gigs['gig_id']
    print(recommended_gigs)
    

user_id='994cbef7-327c-48a9-be42-6e07b4733815'
num_recommendations = 3
content_based_recomendation(str(user_id), num_recommendations)
```

```python
# Collaborative filtering
def collaborative_filtering_recommendation(user_id, num_recommendations):
    # Filter active gigs
    active_gigs = data_gigs[data_gigs['is_active'] == 1]

    # Load dataset (combining view histories, wishlist, and ratings data)
    combined_data = pd.concat([data_view_gigs_history, data_wishlist, data_rating])

    # Create Surprise Dataset
    reader = Reader(rating_scale=(1, 5))
    data = Dataset.load_from_df(combined_data[['user_id', 'gig_id', 'rating']], reader)

    # Split dataset into train set and test set
    trainset, testset = train_test_split(data, test_size=0.2)

    # Build model using KNN algorithm
    model = KNNBasic()

    # Train model with train set
    model.fit(trainset)

    # Predict ratings for gigs that user hasn't interacted with
    user_interacted_gigs = combined_data[combined_data['user_id'] == user_id]['gig_id']
    gigs_to_predict = active_gigs[~active_gigs['gig_id'].isin(user_interacted_gigs)]
    predicted_ratings = []
    for gig_id in gigs_to_predict['gig_id']:
        predicted_rating = model.predict(user_id, gig_id).est
        predicted_ratings.append((gig_id, predicted_rating))

    # Sort predicted ratings and get top recommended gigs
    predicted_ratings.sort(key=lambda x: x[1], reverse=True)
    recommended_gigs = active_gigs.loc[active_gigs['gig_id'].isin([gig_id for gig_id, _ in predicted_ratings[:num_recommendations]])]

    return recommended_gigs

# Example usage
user_id = 1
num_recommendations = 5


# Check if user has performed any activities
#if (user_id in data_view_gigs_history['user_id'].values or 
    #user_id in data_wishlist['user_id'].values or 
    #user_id in data_rating['user_id'].values):
    # Collaborative filtering recommendation
recommendations = collaborative_filtering_recommendation(user_id, num_recommendations)
#else:
    # Content-based filtering recommendation
#recommendations = content_based_recommendation(user_id, num_recommendations)

print(recommendations)
```
