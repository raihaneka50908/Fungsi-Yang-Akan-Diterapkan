```python
import surprise
from surprise import accuracy
```

```python
reader = surprise.Reader(rating_scale=(0,10))
data = surprise.Dataset.load_from_df(df_ratings[['user_id','gig_id','rating']],reader)
```

```python
trainset, testset = surprise.model_selection.train_test_split(data, test_size=0.2)
model = surprise.SVD()
model.fit(trainset)
prediction = model.test(testset)
print(accuracy.mse(prediction))
```

```python
user_id = '99643751-6115-4bbe-b615-5ecea70010e7'
user_gig_images = df_ratings[df_ratings['user_id']==user_id]
gig_to_recomend = df_gig_images[~df_gig_images['gig_id'].isin(user_gig_images)]['gig_id']
prediction = []
for gig_id in gig_to_recomend:
    prediction.append(model.predict(user_id,gig_id))
prediction
```

```python
recommendations = pd.DataFrame(prediction,columns=['uid', 'iid', 'r_ui', 'est', 'details'])
recommendations.sort_values(by='est',ascending=False, inplace=True)
top_5_gigs = recommendations.head(5)['iid'].head(5)
top_5_gigs
```

## Fungsi Untuk Memberikan Rekomendasi

```python
import random
def give_recommendation(user_id):
    user_gig_images = df_ratings[df_ratings['user_id']==user_id]
    gig_to_recomend = df_gig_images[~df_gig_images['gig_id'].isin(user_gig_images)]['gig_id']
    random.shuffle(gig_to_recomend)
    prediction = []
    for gig_id in gig_to_recomend:
        prediction.append(model.predict(user_id,gig_id))
    recommendations = pd.DataFrame(prediction,columns=['uid', 'iid', 'r_ui', 'est', 'details'])
    recommendations.sort_values(by='est',ascending=False, inplace=True)
    top_10_gigs = recommendations.head(10)['iid']
    return top_10_gigs
```

```python
import os
import cv2
import random
import shutil

def hapus_dot_underscore(lokasi, files):
    for index, filee in enumerate(files):
        if filee.startswith('._'):
            os.remove(os.path.join(lokasi, filee))

def ambil_list_gambar(files):
    return [file for _, file in enumerate(os.listdir(files))]

def pindahkan_ke_file_baru(old_location, new_location):
    l_gambar = [file for _, file in enumerate(os.listdir(old_location))]
    jumlah_gambar = len(l_gambar)
    for i in range(0, jumlah_gambar):
        old_file = os.path.join(old_location, l_gambar[i])
        new_file = os.path.join(new_location, l_gambar[i])
        if os.path.exists(new_file):
            os.replace(old_file, new_file)
        else:
            shutil.copy(old_file, new_file)

def rename_files(path, prefix):
    i = 1
    for file_name in os.listdir(path):
        file_path = os.path.join(path, file_name)
        if os.path.isfile(file_path):
            file_ext = "." + file_name.split(".")[-1]
            new_file_name = "{}_{}{}".format(prefix, i, file_ext)
            new_file_path = os.path.join(path, new_file_name)
            if not os.path.exists(new_file_path):
                os.rename(file_path, new_file_path)
                i += 1

def plagiarism_detection(image_path, database_folder_path):
    # Membersihkan file-file gambar yang bermasalah
    files = os.listdir(database_folder_path)
    hapus_dot_underscore(database_folder_path, files)
    
    similarities = []
    selected_images = random.sample(os.listdir(database_folder_path), 10)

    # Membaca gambar1
    image1 = cv2.imread(image_path)
    if image1 is None:
        raise ValueError(f"Gambar tidak dapat dibaca: {image_path}")
    image1 = cv2.resize(image1, (225, 225))

    for filename in selected_images:
        # Membaca gambar dari database
        image2 = cv2.imread(os.path.join(database_folder_path, filename))
        if image2 is None:
            raise ValueError(f"Gambar tidak dapat dibaca: {os.path.join(database_folder_path, filename)}")
        image2 = cv2.resize(image2, (225, 225))

        # Menghitung tingkat kesamaan menggunakan compare_ssim
        similarity = compare_ssim(image1, image2, win_size=3, multichannel=True)

        # Menambahkan nilai kesamaan ke dalam list similarities
        similarities.append(similarity)

    return similarities

# Contoh pemanggilan fungsi plagiarism_detection
image_path = '/Volumes/Kerja_Keras/Folder_Fold_Recovery/For Career/Implementasi_Dengan_Database/Program Percobaan Dengan Database/Percobaan_2_gambar/Gambar1.jpeg'
database_folder_path = 'Data_Gambar/Data_Masuk'

similarities = plagiarism_detection(image_path, database_folder_path)
list_kesamaan = []
# Menampilkan hasil kesamaan
for i, similarity in enumerate(similarities):
    print(f"Kesamaan dengan gambar {i+1}: {similarity}")
    list_kesamaan.append(similarity)
kesamaan = max(list_kesamaan)
print("Gambar Memiliki kesamaan dengan gambar di database sebesar "+str(kesamaan*100)+" Persen")
```
